```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
import matplotlib.pyplot as pp
from sklearn.metrics import adjusted_rand_score
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
x, y = df["x"], df['y']
x.shape, y.shape
```

```python
pp.scatter(
    x=x, y=y, c=df["color"], s=50
)
pp.title("Unclustered data")
```

```python
X = df[["x", "y"]]
X.shape
```

```python
nc = 3
clf = KMeans(n_clusters = nc)
c = clf.fit_predict(X)
```

```python
pp.scatter(
    x=x, y=y, c=c, s=50
)
pp.title(f"Clustered data ({nc} clusters)")
```

```python
score = adjusted_rand_score(df['color'], c)
print(f"   : {score:.3f}")
```
